```python
# Задание task_03_01_07.
#
# Выполнил: Гришина А.А.
# Группа: ЦИБ-251

def has_digits(предложение):
    """Проверяет, содержит ли предложение хотя бы одну цифру."""
    for char in предложение:
        if char.isdigit():
            return True
    return False

def количество_предложений_c_цифрами(предложения):
    """Подсчитывает количество предложений, содержащих хотя бы одну цифру."""
    count = 0
    for предложение in предложения:
        if has_digits(предложение):
            count += 1
    return count

n = int(input("Введите количество предложений: "))

предложения = []
for i in range(n):
    print(f"Введите предложение №{i + 1}:")
    text = input()
    предложения.append(text)

result = количество_предложений_c_цифрами(предложения)
print(f"Предложений с цифрой = {result}")
# ---
# Пример вывода:
#
# Введите количество предложений: 3
# Введите предложение №1:
# Просто текст
# Введите предложение №2:
# Текст с цифрой 1 (один)
# Введите предложение №3:
# Тут нет цифры
# Предложений с цифрой = 1
```
